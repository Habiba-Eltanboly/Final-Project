### 1- what is the total sales in the store?

```python
total_sales = df['Sales'].sum()
print(f'Total Sales = {total_sales}')
```

#### Total sales by year

```python
sales_by_year = (
    df.groupby(df['Order Date'].dt.year)['Sales']
    .sum()
    .round(2)
    .reset_index(name='TotalSales')
    .rename(columns={'Order Date': 'Year'})
    .sort_values('Year')
)
sales_by_year
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sales_by_year['Year'], sales_by_year['TotalSales'], color='purple')
plt.title('Total Sales by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Total sales by m

```python
monthly_sales = (
    df.groupby([df['Order Date'].dt.year.rename('Year'), 
                df['Order Date'].dt.month.rename('Month')])['Sales']
    .sum()
    .round(2)
    .reset_index(name='MonthlySales')
    .sort_values(['Year', 'Month'])
)
monthly_sales
```

```python
plt.figure(figsize=(14, 6))
x_labels = monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str).str.zfill(2)
x_values = range(len(monthly_sales))
plt.plot(x_values, monthly_sales['MonthlySales'], marker='o', color='#1f77b4', linewidth=2)
plt.xticks(ticks=x_values, labels=x_labels, rotation=45, fontsize=10)
for i, label in enumerate(plt.gca().xaxis.get_ticklabels()):
       if i % 3 != 0:  # Changed from n to 3 for clarity
           label.set_visible(False)
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 2- which region has the highest sales?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=region_sales, x="Region", y="Sales", hue='Region', palette="viridis")
plt.title("Total Sales by Region")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.show()
```

### 3- which city has the highest number of orders?

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_cities_sales, x="City", y="Sales", hue='City', palette="magma")
plt.title("Total Sales by City")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.show()
```

### 4- which category has the highest sales?

```python
category_sales = df.groupby("Category")["Sales"].sum().reset_index()
category_sales
```

```python
plt.figure(figsize=(8, 8))
plt.pie(category_sales["Sales"], labels=category_sales["Category"], autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Sales Distribution by Product Category")
plt.show()
```

### 5-what are the top Selling products?

```python
top_10_products = (
    df.groupby('Product Name')
    .agg(
        TotalSales=('Sales', 'sum'),
        OrderCount=('Sales', 'count')
    )
    .round(2)
    .sort_values('TotalSales', ascending=False)
    .head(10)
    .reset_index()
)

top_10_products
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=top_10_products,
    x="TotalSales",
    y="Product Name",
    palette="viridis",
    hue="Product Name",
    legend=False,
    dodge=False
)
plt.title("Top 10 Products by Sales", fontsize=16, pad=20)
plt.xlabel("Total Sales ($)", fontsize=12)
plt.ylabel("")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
for p in ax.patches:
    width = p.get_width()
    ax.text(
        width + (0.01 * top_10_products["TotalSales"].max()),  # Small offset
        p.get_y() + p.get_height() / 2,
        f"${width:,.2f}",
        va="center",
        ha="left",
        fontsize=10
    )
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```

### 7-which shipping mode is most commonly used?

```python
ship_mode_counts = df["Ship Mode"].value_counts().reset_index()
ship_mode_counts.columns = ["Ship Mode", "Order Count"]
ship_mode_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    ship_mode_counts["Order Count"],
    labels=ship_mode_counts["Ship Mode"],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("viridis"),
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)
plt.title("Order Distribution by Shipping Mode", pad=20)
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

### 8-what is the average time between order date and ship date?

```python
avg_delivery_days = (df['Ship Date'] - df['Order Date']).dt.days.mean()

print(f"Average delivery days: {avg_delivery_days:.2f}")
```

### 9-are there delays in time (order to ship) in certain regions?

```python
avg_delivery_by_region = (
    df.assign(DeliveryDays=(df['Ship Date'] - df['Order Date']).dt.days)
    .groupby('Region')['DeliveryDays']
    .mean()
    .reset_index()
    .rename(columns={'DeliveryDays': 'AvgDeliveryDays'})
    .sort_values('AvgDeliveryDays', ascending=False)
)

avg_delivery_by_region
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=avg_delivery_by_region,
    y='Region',
    x='AvgDeliveryDays',
    hue='Region',
    palette="magma",
    order=avg_delivery_by_region.sort_values('AvgDeliveryDays', ascending=False)['Region']
)
plt.title('Average Delivery Days by Region')
plt.xlabel('Days')
plt.ylabel('')
sns.despine()
plt.tight_layout()
plt.show()
```

### 10-how does shipping mode affect delivery time?

```python
ship_mode_delivery = (
    df.assign(DeliveryDays=(df['Ship Date'] - df['Order Date']).dt.days)
    .groupby('Ship Mode')
    .agg(AvgDeliveryDays=('DeliveryDays', 'mean'),
         OrderCount=('DeliveryDays', 'count'))
    .reset_index()
    .sort_values('AvgDeliveryDays')
)

ship_mode_delivery
```

```python
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=ship_mode_delivery,
    x='AvgDeliveryDays',
    y='Ship Mode',
    size='OrderCount',
    sizes=(100, 1000),
    hue='Ship Mode',
    palette="viridis",
    legend=False
)
sns.lineplot(
    data=ship_mode_delivery,
    x='AvgDeliveryDays',
    y='Ship Mode',
    color='gray',
    alpha=0.3
)

for i, row in ship_mode_delivery.iterrows():
    plt.text(
        row['AvgDeliveryDays'] + 0.3,
        i,
        f"{row['AvgDeliveryDays']:.1f} days\n({row['OrderCount']:,} orders)",
        va='center',
        fontsize=10
    )
plt.title('Shipping Mode vs Delivery Time', fontsize=16)
plt.xlabel('Average Delivery Days')
plt.ylabel('Shipping Mode')
sns.despine()
plt.tight_layout()
plt.show()
```
